```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
file_path = "/content/university_student.csv"
df = pd.read_csv(file_path)
```

```python
display(df.head())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Gender', 'School Type'], drop_first=True)
```

```python
label_encoder = LabelEncoder()
df['Grade'] = label_encoder.fit_transform(df['Grade'])
```

```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df['Age'] = (pd.to_datetime('today') - df['Date of Birth']).dt.days // 365
df['Total Score'] = df[['Math Score', 'Reading Score', 'Writing Score']].sum(axis=1)
```

```python
grouped_data = df.groupby('Grade').agg({'Total Score': 'mean', 'Study Time': 'mean', 'Absences': 'sum'})

```
